```python
df = pd.read_csv('skindataall.csv', index_col=[0])
```

```python
def create_interaction_matrix(df, user_col, item_col, rating_col, norm= False, threshold =0):
    interactions = df.groupby([user_col, item_col])[rating_col].sum().unstack().reset_index().fillna(1).set_index(user_col)
    if norm:
        interactions = interactions.applymap(lambda x: 1 if x > threshold else 0)
    return interactions
```

```python
X = df.Ingredients
y = df.Category
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 42)
categories = ['Moisturizer', 'Cleanser', 'Face Mask', 'Treatment']
#Logistic Regression
logreg = Pipeline([('vect', CountVectorizer()),
                ('tfidf', TfidfTransformer()),
                ('clf', LogisticRegression(n_jobs=1, C=1e5)),
               ])
logreg.fit(X_train, y_train)


y_pred = logreg.predict(X_test)

print('accuracy %s' % accuracy_score(y_pred, y_test))
print(classification_report(y_test, y_pred,target_names=categories))
```
